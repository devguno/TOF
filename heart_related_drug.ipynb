```python
import pandas as pd

# Load the data
file_path = 'drug_count.csv'
drug_data = pd.read_csv(file_path)

# Check for and handle any null values in the dataset
drug_data_cleaned = drug_data.dropna()

# Summarize total prescriptions per drug
prescriptions_per_drug = drug_data_cleaned.groupby('drug')['count'].sum().reset_index()
prescriptions_per_drug_sorted = prescriptions_per_drug.sort_values('count', ascending=False)

# Show the summarized data
prescriptions_per_drug_sorted
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data preparation for visualization
division_prescriptions = drug_data_cleaned.groupby(['division', 'drug'])['count'].sum().unstack().fillna(0)

# Enhancing the bar chart with counts displayed on top
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(12, 8))

# Create the stacked bar chart
division_prescriptions.plot(kind='bar', stacked=True, colormap='coolwarm', ax=ax)

# Adding drug names and count labels on each bar
for rec, (drug, div) in zip(ax.patches, division_prescriptions.unstack().index):
    height = rec.get_height()
    y = rec.get_y()
    if height > 0 and div != "etc":  # Only display labels for non-zero segments
        label_text = f"{drug}: {int(height)}"
        ax.text(rec.get_x() + rec.get_width() / 2, y + height / 2, label_text, ha='center', va='center', color='black', fontsize=10, fontweight='bold')

plt.title('Distribution of Drug Prescriptions Across Different Divisions')
plt.xlabel('Division')
plt.ylabel('Total Prescriptions')
plt.xticks(rotation=45)
plt.legend(title='Drug', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()

```

```python
import pandas as pd

df = pd.read_csv('tofdrug2_ko.csv', encoding='utf-8')
df.head()
```

```python
drug_keywords = [
    "FRS4", "SRN2", "HCT", "ENL", "ENL5", "LST100", "LST25", "CVD3", "CVD6", "CVD1", "CVD", "CVDS8", 
    "CVDS16", "CVDS64", "CVDS32", "ASAT1", "ASAE", "ASA500", "CPD", "CPDN", "WFR2", "WFR5", 
    "SCBVS50", "SCBVS100", "SCBVS200", "TVT15", "TVT30", "DPGZ"
]
```

```python
new_drug_keywords = [
    "FCND50", "QND", "MXLT", "LDC2I", "LDC1I", "LDC2I5", "NDL", "ATN", "ATN2", "PPN1", "PPN4", "PPNS",
    "MTP", "BSPL", "BSPL1K", "BSPL2", "BSPL10", "CVD3", "CVD6", "CVD1", "CVD", "CVDS8", "CVDS16",
    "CVDS64", "CVDS32", "AMO", "AMOI", "STL4", "VRP18", "VRP4", "VRP8", "VRPI", "DTA3IR", "DTA3",
    "DTA9", "DTAI", "DTA18", "DGXL", "DGX25", "DGXI", "ADN90I", "ADNI"
]
```

```python
# Attempting the analysis again by creating a summary table for each drug keyword in drug_keywords
results = []

# Ensure 'drug_source_value' exists in the columns
if 'drug_source_value' in df.columns:
    for keyword in new_drug_keywords:
        # Filter df for each keyword
        specific_drug_data = df[df['drug_source_value'] == keyword]
        
        # Drop duplicates based on 'person_id' to ensure unique persons are considered for gender distribution
        unique_person_data = specific_drug_data.drop_duplicates(subset='person_id')
        
        # Calculate unique person count
        unique_persons = unique_person_data['person_id'].nunique()
        
        # Gender distribution from the unique persons df
        gender_dist = unique_person_data['gender_source_value'].value_counts().to_dict()
        
        # Append results
        results.append({
            'Drug Keyword': keyword,
            'Unique Person Count': unique_persons,
            'Gender Distribution': gender_dist
        })

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame(results)
results_df
```
